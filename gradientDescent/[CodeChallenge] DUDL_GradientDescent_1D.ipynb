### STEP 1

```
import sympy as sym
import sympy.plotting.plot as symplot
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
from IPython.display import display
from sympy import *
```

```
x = sym.symbols('x')
cos = sym.symbols('cos')
pi = sym.symbols('π')
fx = sym.cos(2*sym.pi*x) + x**2 # function
df = sym.diff(fx) # derivative of the function
```

```
print('The functions:')

display(fx)
display(df)
```

```
# sympyの数式を関数化できちゃう！
lambdify??
```

```
fx = lambdify(x,fx)
df = lambdify(x,df)
```

```
#  plot the function and its derivative
# define a range for x
x_labels = np.linspace(-2,2,2001)
# plotting
plt.plot(x_labels ,fx(x_labels), x_labels ,df(x_labels))
# plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```
x_labels
```

```
# random starting point
localmin = np.random.choice(x_labels,1)
print(f"starting point of x: {localmin}")

# learning parameters
learning_rate = 0.01
training_epochs = 100000

# run through training
for i in range(training_epochs):
  grad = df(localmin)
  localmin = localmin - learning_rate*grad

print(f"x in the result of learning: {localmin}")
```

```
# plot the results

plt.plot(x_labels ,fx(x_labels), x_labels ,df(x_labels))

plt.plot(localmin,df(localmin),'ro')
plt.plot(localmin,fx(localmin),'bo')

# plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()
```
